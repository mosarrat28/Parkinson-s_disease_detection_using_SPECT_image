##Copying from one folder to another

```python
import os
import shutil

source_folder = "/content/drive/MyDrive/Project/Original_dataset_nonpd/"
destination_folder = "/content/drive/MyDrive/Project/Original_dataset2/nonpd/"

# fetch all files
for file_name in os.listdir(source_folder):
    # construct full file path
    source = source_folder + file_name
    destination = destination_folder + file_name
    # copy only files
    if os.path.isfile(source):
        shutil.copy(source, destination)
        print('copied', file_name)
    else:
        print('not copied')
```
